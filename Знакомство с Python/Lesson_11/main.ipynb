## Инициализируем функцию ## 

```python
from sympy import *

x, f = symbols('x f')
f = -12*x**4 - 18*x**3 + 5*x**2 + 10*x - 30
f
```

```python
plot(f)
```

```python
limit_expr = limit(f, x, 1-oo)

limit_expr
```

```python
limit_expr = limit(f, x, oo)

limit_expr
```

```python
plot(f,(x,-2.5,2.5))
```

 ## Определение корней ## 

```python
solveset(f, x) # Определить корни
```

Функция имеет корни лишь в комплексной плоскости

## Интервалы возрастания и убывания ## 

```python
# Для этого нужно взять производную f'(x)
d_f = f.diff(x) 
d_f
```

```python
plot(d_f)
```

```python
plot(d_f,(x,-1.5,1.5))
```

На интервалах, где производная >0, функция возрастает.

На интервалах, где производная <0, функция убывает.

Для начала найдём стационарные точки


```python
solve(d_f)
```
