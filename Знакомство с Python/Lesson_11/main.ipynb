```python
from sympy import *


# f(x) = -12x^4*sin(cos(x)) - 18x^3+5x^2 + 10x - 30
# Определить корни
# Найти интервалы, на которых функция возрастает
# Найти интервалы, на которых функция убывает
# Построить график
# Вычислить вершину
# Определить промежутки, на котором f > 0
# Определить промежутки, на котором f < 0

x, f = symbols('x f')
f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30
f
```

```python
solve(f, (x, -11,-5)) # Определить корни
```

```python
d_f = f.diff(x) #Найти интервалы, на которых функция возрастает
plot(d_f,x)
```

```python
plot(f,x) # Построить график
```
